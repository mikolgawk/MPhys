```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.preprocessing import image
from tensorflow.keras.optimizers import RMSprop
import matplotlib.pyplot as plt
import tensorflow as tf
import numpy as np
import cv2
import os

img = image.load_img("/home/mikolaj/DataTraining/1/2dm-2500.png")
plt.imshow(img)
dim1=200
dim2=200
```

```python
cv2.imread("/home/mikolaj/DataTraining/1/2dm-2500.png").shape
```

```python
train = ImageDataGenerator(rescale=1/255)
validation = ImageDataGenerator(rescale=1/255)
train_dataset = train.flow_from_directory('/home/mikolaj/DataTraining/',target_size=(dim1, dim2), batch_size = 3, class_mode = 'binary')
```

```python
validation_dataset = train.flow_from_directory('/home/mikolaj/DataValidation/',target_size=(dim1, dim2), batch_size = 3, class_mode = 'binary')
```

```python
train_dataset.class_indices
```

```python
train_dataset.classes
```

```python
model = tf.keras.models.Sequential([tf.keras.layers.Conv2D(16, (3, 3), activation = 'relu', input_shape = (dim1, dim2, 3)),
                                   tf.keras.layers.MaxPool2D(2, 2),
                                   tf.keras.layers.Conv2D(32, (3, 3), activation = 'relu', input_shape = (dim1, dim2, 3)),
                                   tf.keras.layers.MaxPool2D(2, 2),
                                   tf.keras.layers.Conv2D(64, (3, 3), activation = 'relu', input_shape = (dim1, dim2, 3)),
                                   tf.keras.layers.MaxPool2D(2, 2), tf.keras.layers.Flatten(), tf.keras.layers.Dense(512,activation='relu'), tf.keras.layers.Dense(1,activation='sigmoid')])
```

```python
model.compile(loss='binary_crossentropy',optimizer=RMSprop(lr=0.001),metrics=['accuracy'])
```

```python
model_fit = model.fit(train_dataset, steps_per_epoch=5, epochs=30, validation_data=validation_dataset)
```
